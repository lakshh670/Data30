```python
# let's perform statistical tests on tips dataset
# let's first seperate the tips by time
from scipy import stats
lunch_tips = tips[tips['time'] == 'Lunch']['tip']
dinner_tips = tips[tips['time'] == 'Dinner']['tip']
```

```python
print("Lunch mean tip:", np.mean(lunch_tips))
print("Dinner mean tip:", np.mean(dinner_tips))
```

```python
#Now let's perform an independent t-test
t_stat, p_value = stats.ttest_ind(lunch_tips, dinner_tips)
print("\nT-statistic:", t_stat)
print("P-value:", p_value)
# Interpretation
if p_value < 0.05:
    print("Reject H0: Mean tips differ significantly between Lunch and Dinner")
else:
    print("Fail to reject H0: No significant difference in mean tips")
```

```python
# Let's compute correlation between total_bill and tip
corr_coef, p_value = stats.pearsonr(tips['total_bill'], tips['tip'])
print("Pearson correlation coefficient:", corr_coef)
print("P-value:", p_value)
# Interpretation
if p_value < 0.05:
    print("Reject H0: Total bill and tip are significantly correlated")
else:
    print("Fail to reject H0: No significant correlation")
```
